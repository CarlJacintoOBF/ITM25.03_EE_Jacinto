````python
import csv #list of values that need not be cleaned (first name, last name and birth day)

first_name_list= []
last_name_list=[]
birth_day_list=[]
f= open('recweekdata.csv')
csv_f= csv.reader(f, delimiter= '	')

for row in csv_f:
    first_name_list.append(row[0])
    last_name_list.append(row[1])
    birth_day_list.append(row[3])
    
    
````

````python
last_name_list.remove('lastname') #cleaning up headers
````

````python
birth_day_list.remove('birthday') #cleaning up headers
````

````python
birthday_tally={} #tallying birthday data

bd_set= set(birth_day_list)

for day in bd_set:
    birthday_tally[day]=0
    
for day in birth_day_list:
    birthday_tally[day] +=1
    
birthday_tally
````

````python
import csv #imported birth month data

birth_month_list= []
f= open('recweekdata.csv')
csv_f= csv.reader(f, delimiter= '	')

for row in csv_f:
    birth_month_list.append(row[2])
````

````python
bm_tally= {} #tallying birth month data

bm_set= set(clean_bm_list)

for month in bm_set:
    bm_tally[month]=0
    
for month in clean_bm_list:
    bm_tally[month] +=1
    
bm_tally
````

````python
import csv #importing cellphone data
 
cpno_list= []
f= open('recweekdata.csv')
csv_f= csv.reader(f, delimiter= '	')


for row in csv_f:
    cpno_list.append(row[6])


    
````

````python
un_cpno_list=list(set(cpno_list)) #converting to list unique values
````

````python
un_cpno_list #checking
````

````python
import re #replacing values to uniform format
uniform_no_list=[]

for num in un_cpno_list:
    replaced_num= re.sub(r"^9([0-9][0-9])", r"+63 9\1", num)
    uniform_no_list.append(replaced_num)
    
    
````

````python
uniform_no_list #checking
````

````python
network_tally= {"Smart":0, "Globe":0} #tallying network data

    
for num in uniform_no_list:
    globe= re.search(r'[+]63\s9[1-2]7\s[0-9]{3}\s[0-9]{4}', num)
    smart= re.search(r'[+]63\s918\s[0-9]{3}\s[0-9]{4}', num)
    if smart:
        network_tally["Smart"]+=1
    if globe:
        network_tally["Globe"]+=1
        
````
